```python

%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
x = np.array([])
y = np.array([])
```

```python
a = (0,0)
b = (5,10)
c = (10,0)

abc = [a,b,c]
abc[0][0]
```

```python
midpoint =  ((abc[0][0]+abc[1][0]/2),(abc[0][1]+abc[1][1]/2))
midpoint
```

```python
for i in range(20000):
    point = random.choice(abc)
    midpoint = ((midpoint[0]+point[0])/2,(midpoint[1]+point[1])/2)
    x = np.append(x,midpoint[0])
    y = np.append(y,midpoint[1])
    
```

```python
plt.scatter(x,y, marker="x", s=0.1)
plt.show()
```
