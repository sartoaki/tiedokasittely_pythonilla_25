```python
import numpy as np
```

```python
weights = [80, 90, 45, 60, 120, 99, 73, 82, 57, 63]
heights = [180, 176, 155, 170, 179, 175, 190, 160, 169, 169]
```

```python
w_array = np.array(weights)
h_array = np.array(heights)
```

```python
w_array
```

```python
h_array
```

```python
for i in range(len(w_array)):
    bmi = w_array[i] / (h_array[i] / 100)**2
    print("Weight: ", w_array[i], "Height: ", h_array[i], "BMI: ", bmi)
```
